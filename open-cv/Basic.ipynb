## Capture properties

```python
#height
height = capture.get(cv2.CAP_PROP_FRAME_HEIGHT)
```

```python
#width
width = capture.get(cv2.CAP_PROP_FRAME_WIDTH)
```

```python
#Get number of frame
capture.get(cv2.CAP_PROP_FRAME_COUNT)
```

```python
#Get number of frame per second
fps = capture.get(cv2.CAP_PROP_FPS)
```

## Writing our videos

```python
video_path = os.path.join('data','video','family_1.mp4')
capture = cv2.VideoCapture(video_path)

output_video = os.path.join('data','video','output_1.avi')

video_writer = cv2.VideoWriter(output_video
                              ,cv2.VideoWriter_fourcc('P','I','M','1')
                              ,fps
                              ,(int(width),int(height))
                               ,isColor=FALSE)

for frame_index in range(int(capture.get(cv2.CAP_PROP_FRAME_COUNT))):
    ret, frame = capture.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('video player',frame)
    video_writer.write(gray)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()

video_writer.release()
```
